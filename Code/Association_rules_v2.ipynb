# Apriori for Mobile Services

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder

data = pd.read_csv('/content/transactions.csv')

# Step 1: Filter the data to only include rows where serviceType is 'Mobile Services'
mobile_services_data = data[data['serviceType'] == 'Mobile Services']

# Step 2: Create a list of transactions (invoiceID -> list of items bought)
transactions = mobile_services_data.groupby('invoiceID')['serviceSubType'].apply(list).values.tolist()

# Step 3: Convert data to a format suitable for apriori
encoder = TransactionEncoder()
encoded_data = encoder.fit_transform(transactions)

# Step 4: Create a DataFrame of the transformed data
df = pd.DataFrame(encoded_data, columns=encoder.columns_)

# Step 5: Apply the apriori algorithm to find frequent itemsets
frequent_itemsets = apriori(df, min_support=0.1, use_colnames=True)

# Step 6: Generate the association rules using the 'lift' metric
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

# Step 7: Visualize or inspect the rules
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```

# FP Growth for Streaming Services

```python
import pandas as pd
from mlxtend.frequent_patterns import fpgrowth, association_rules
from mlxtend.preprocessing import TransactionEncoder

# Load your dataset
data = pd.read_csv('/content/transactions.csv')

# Step 1: Filter data to only include rows where serviceType is 'streaming Services'
enterprise_services_data = data[data['serviceType'] == 'Streaming Services']

# Step 2: Create a list of transactions (invoiceID -> list of items bought)
transactions = enterprise_services_data.groupby('invoiceID')['serviceSubType'].apply(list).values.tolist()

# Step 3: Convert data to a format suitable for FP-Growth
encoder = TransactionEncoder()
encoded_data = encoder.fit_transform(transactions)

# Step 4: Create a DataFrame of the transformed data
df = pd.DataFrame(encoded_data, columns=encoder.columns_)

# Step 5: Apply the FP-Growth algorithm to find frequent itemsets
frequent_itemsets = fpgrowth(df, min_support=0.1, use_colnames=True)

# Step 6: Generate the association rules using the 'lift' metric
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

# Step 7: Visualize or inspect the rules
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```
